# Visualización de tablas

```python
import pandas as pd

# Crear un DataFrame con los datos
datos = {
    'Nombre': ['Ana', 'Juan', 'Sofía', 'Carlos', 'Valentina'],
    'Matemáticas': [85, 88, 95, 70, 80],
    'Ciencias': [90, 76, 89, 85, 87],
    'Historia': [78, 85, 92, 75, 88],
    'Inglés': [92, 80, 91, 82, 90]
}

df = pd.DataFrame(datos)

df
```

# Creación de gráficos

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 3))

# Graficar los datos
for materia in ['Matemáticas', 'Ciencias', 'Historia', 'Inglés']:
    plt.plot(df['Nombre'], df[materia], marker='o', label=materia)

# Personalizar el gráfico
plt.title('Rendimiento de los Estudiantes por Materia', fontsize=16)
plt.xlabel('Estudiantes', fontsize=12)
plt.ylabel('Puntajes', fontsize=12)
plt.legend(title='Materias')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar el gráfico
plt.show()
```

# Juego sencillo

```python
import random

MENSAJES = {
    0: "Ganaste!",
    1: "El número debe ser menor",
   -1: "El número debe ser mayor",
}

def comparar_numeros(numero_secreto: int, intento: int) -> int:
    if intento > numero_secreto:
        return 1
    elif intento < numero_secreto:
        return -1
    else:
        return 0 # son iguales

def jugar():
    numero_secreto = random.randint(1, 100)
    while True:
        intento = int(input("Ingresa el intento: "))
        resultado = comparar_numeros(numero_secreto, intento)
        print(MENSAJES[resultado])
        if resultado == 0:
            break

jugar()
```

# Colab Turtle

```python
!pip install ColabTurtle
import random
from ColabTurtle.Turtle import *

initializeTurtle()
speed(10)

def estrella(largo):
  for _ in range(5):
    forward(largo)
    right(144)

def mover_al_azar():
  penup()
  x = random.randint(0, 800)
  y = random.randint(0, 500)
  goto(x, y)
  pendown()

def rgb():
  r = random.randint(0, 255)
  g = random.randint(0, 255)
  b = random.randint(0, 255)
  return r, g, b

def cielo_estrellado(n):
  for _ in range(n):
    color(rgb())
    estrella(random.randint(3, 10))
    mover_al_azar()

cielo_estrellado(40)
hideturtle()
```

# Animación simple

```python
import time

for char in "Hola, Mundo!":
    print(char, end='', flush=True)
    time.sleep(0.2)
time.sleep(1)

# Ahora borramos la palabra "Mundo!"
for _ in range(6):  # El tamaño de la palabra "Mundo!"
    print('\b', end='', flush=True)  #  chr(8)  chr(0x08)  chr(0b0000100)  '\b'  '\x08'
    time.sleep(0.2)
time.sleep(0.5)

# Escribimos la nueva palabra
print("Python!", end='', flush=True)
```
